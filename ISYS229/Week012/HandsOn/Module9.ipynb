```python
import json
import random
import tkinter as tk
from ref.random_quote_generator import get_random_quote


def divide_numbers():
    try:
        num1 = float(num1_entry.get())
        num2 = float(num2_entry.get())
        result = num1/num2
        result_label.config(text=f"Result: {result}")
    except ValueError:
        result_label.config(text="Please enter valid numbers")
    except ZeroDivisionError:
        result_label.config(text="Cannot divide by zero")
    finally:
        num1_entry.delete(0, tk.END)
        num2_entry.delete(0, tk.END)


def double_number():
    try:
        num = int(num1_entry.get())
        result = num*2
        result_label.config(text=f"Result: {result}")
    except ValueError:
        result_label.config(text="Please enter a valid integer")
    finally:
        num1_entry.delete(0, tk.END)


def generate_quote():
    quote_text = get_random_quote(wordcount=20)
    quote_label.config(text=quote_text)


root = tk.Tk()
root.title("Why MATH...?")
num1_label = tk.Label(root, text="Enter first number:")
num1_label.grid(row=1, column=0)
num1_entry = tk.Entry(root)
num1_entry.grid(row=1, column=1)

num2_label = tk.Label(root, text="Enter second number:")
num2_label.grid(row=2, column=0)
num2_entry = tk.Entry(root)
num2_entry.grid(row=2, column=1)

divide_button = tk.Button(root, text="Divide", command=divide_numbers)
divide_button.grid(row=3, column=0)

double_button = tk.Button(root, text="Double First #", command=double_number)
double_button.grid(row=3, column=1)

result_label = tk.Label(root, text="")
result_label.grid(row=4, column=0, columnspan=2)

quote_button = tk.Button(root, text="Generate Quote", command=generate_quote)
quote_button.grid(row=5, column=0)

quote_label = tk.Label(root, text="")
quote_label.grid(row=6, column=0, columnspan=2)


root.mainloop()
```
